```python
page = requests.get('https://www.karar.com/yazarlar/zafer-acar')
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
links = soup.find_all('article', {'class': 'item box-shadow'})
```

```python
links
```

```python
posts = []
titles = []
for link in links:
    post_link = 'https://www.karar.com'+link.a.get('href')
    title = link.h3.text
    post_page = requests.get(post_link)
    post_soup = BeautifulSoup(post_page.content, 'html.parser')
    article = post_soup.select('article.article-content')[0].text

    posts.append(article)
    titles.append(title)   
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'title': titles, 'post': posts})
```

```python
df
```

```python
df['post'][0]
```

```python
df['post'] = df['post'].str.replace('\n', ' ')  
```
