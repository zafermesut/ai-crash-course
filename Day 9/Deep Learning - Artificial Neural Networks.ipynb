### Deep Neural Networks Classification

```python
import pandas as pd 
```

```python
df = pd.read_csv('pima-indians-diabetes.csv')
```

```python
df.head()
```

```python
df.isnull().sum()   
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = df.drop('Outcome', axis=1)
y = df[['Outcome']]
```

```python
model = Sequential()
model.add(Dense(120, activation='relu'))
model.add(Dense(80, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=250, batch_size=32, validation_split=0.2, verbose=1) 
```

```python
model.summary()
```

```python
from sklearn.preprocessing import normalize, scale
```

```python
x = scale(x)
```

```python
history = model.fit(x, y, epochs=250, batch_size=32, validation_split=0.2, verbose=1) 
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['accuracy'], label='Accuracy')
plt.plot(history.history['val_accuracy'], label='Val_Accuracy')
plt.legend();
```

### DNN - Regression 

```python
import pandas as pd
```

```python
df = pd.read_pickle('kc_house.pkl')
```

```python
df.head()
```

```python
x = df.drop('price', axis=1)
y = df[['price']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = Sequential()
model.add(Dense(120, activation='relu'))
model.add(Dense(80, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(20, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
history = model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=1500, batch_size=128)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
r2_score(y_test, pred)
```

```python
mean_squared_error(y_test, pred) ** 0.5
```

```python
loss_f = pd.DataFrame(history.history)
```

```python
loss_f.plot()
```
