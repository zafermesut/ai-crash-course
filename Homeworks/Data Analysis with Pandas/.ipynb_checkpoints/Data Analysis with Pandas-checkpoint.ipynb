#### Reading in Data

```python
import pandas as pd
```

```python
# reading data;
df1 = pd.read_csv('client_list.csv')
df2 = pd.read_table('client_list.table')
df3 = pd.read_fwf('client_list.txt')
```

```python
pd.read_csv('client_list.csv', skiprows=3, header=None) # skipping 3 rows and ignoring header
```

```python
df = df1
df.columns = df.columns.str.upper() # setting column headings to all caps
df.head()
```

```python
pd.read_csv('client_list_practice.csv', usecols=['FIRST_NAME','AGE','EYE_COLOR']) # extracting desired columns
```

### Slicing a Data Set

```python
# Slicing rows 5 through 11
df[4:11] #way-1
df.loc[4:10: 1] #way-2
```

```python
# Returning only the columns ['LAST_NAME','AGE','HAIR_COLOR']
df[['LAST_NAME','AGE','HAIR_COLOR']] #way-1
df.loc[:, ['LAST_NAME','AGE','HAIR_COLOR']] #way-2
```

```python
# combining problems
df[['LAST_NAME','AGE','HAIR_COLOR']][4:11] #way-1
df.loc[4:10:1, ['LAST_NAME','AGE','HAIR_COLOR']] #way-2
```

### Simple Queries

```python
# where the client's last name is "Smith".
df[df['LAST_NAME']=='Smith']
```

```python
# where the client's hair color is not black.
df[df['HAIR_COLOR'] != 'black']
```

```python
# where the client's hair color is red and reset the values to "ginger".
df.loc[df.HAIR_COLOR=='red', 'HAIR_COLOR'] = "ginger" # I got help from github
df.head()
```

### Complex Queries

```python
# where the clients are females older than 30 years.
```

```python
df[(df.AGE>30) & (df.GENDER=='F')]
```

```python
# Repeat problem 1, but return only the hair color and eye color.
df[(df.AGE>30) & (df.GENDER=='F')][['HAIR_COLOR', 'EYE_COLOR']]
```

```python
#Â Find the unique combination of hair and eye color for women older than 25 years.
df[(df.AGE>25) & (df.GENDER=='F')][['HAIR_COLOR', 'EYE_COLOR']].drop_duplicates()
```

### Additional Dataframe Operations

```python
# Performing a merge using "client_list.csv" and "customer_id_list.csv".
ids = pd.read_csv('customer_id_list.csv')
df = pd.read_csv('client_list.csv')
df.columns = df.columns.str.upper()
clients = pd.merge(left=df, right=ids, how='left', on=['LAST_NAME','FIRST_NAME','GENDER','AGE'])
clients.head()
```

```python
# Perform a merge using clients and "purchase_log.csv" 
# and limit the subset to only clients who made purchases. Assign the resulting dataframe as detailed_sales.
purchases = pd.read_csv('purchase_log.csv')
detailed_sales = pd.merge(right=clients, left=purchases, how='inner', on=['CUSTOMER_ID'])
detailed_sales.head(12)
```

```python
#Â Use groupby to find the client who spent the most money on purchases. Determine how much he/she spent
spenders = detailed_sales.groupby(['CUSTOMER_ID','FIRST_NAME','LAST_NAME','GENDER','AGE'], as_index=False)['PRICE'].sum()
spenders[spenders.PRICE==spenders.PRICE.max()]
```

```python
# Bonus
top_spender_id = spenders[spenders.PRICE==spenders.PRICE.max()].reset_index(drop=True).loc[0,'CUSTOMER_ID']
purchases.loc[purchases.CUSTOMER_ID==top_spender_id, 'ITEM_DESCRIPTION']
```

#### Writing Files


```python
# saving detailed_sales as a 'df_out.csv
detailed_sales.to_csv('df_out.csv', index=False)
```

```python
# Saving detailed_sales to "df_out.p"
detailed_sales.to_pickle('df_out.p')
```
